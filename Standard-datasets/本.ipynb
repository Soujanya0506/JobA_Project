```python
# central limit theorem
# z (myu^ - myu)/sigma/root (n)
N = 2000
y = np.zeros(N)
for n in [5,5000]:
  for i in range(N):
    x = scipy.stats.uniform.rvs(size = n)
    y[i] = (x.mean()-1/2)/(np.sqrt(1/12)/np.sqrt(n))
  plt.hist(y,bins=20, range=(-4,4),density = True)
```
