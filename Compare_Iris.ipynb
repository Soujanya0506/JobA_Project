#Scikit-Learn

```python
from sklearn.datasets import load_iris

iris2 = load_iris()

df2 = pd.DataFrame(iris2.data, columns=iris2.feature_names)

df2 = df2.assign(target=iris2.target)
```

```python
df2
```

```python
df2['target'] = df2['target'].replace({0: 'setosa', 1: 'versicolor', 2:'virginica'})
```

```python
# Convert the DataFrame to a CSV file
df2.to_csv('iris_data.csv', index=False)
```

```python
import seaborn as sns

g = sns.pairplot(df2, hue = 'target')
g._legend.remove()
g.fig.legend(
    handles = g._legend_data.values(),
    labels = g._legend_data.keys(),
    loc = 'lower center', ncol = 3,
    title = 'species')
g.fig.subplots_adjust(bottom = 0.12)
```

#RandomForestClassifier Scikit-Learn

```python
#Import train_test_split function
from sklearn.model_selection import train_test_split

X=df2[['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)', 'petal width (cm)']]  # Features
y=df2['target']  # Labels

# Split dataset into training set and test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2) # 80% training and 20% test
```

```python
#Import Random Forest Model
from sklearn.ensemble import RandomForestClassifier

#Create a Gaussian Classifier
clf=RandomForestClassifier(n_estimators=100, max_depth = 5, criterion = 'entropy', max_features = 'log2', max_leaf_nodes = 8, random_state = 1)

#Train the model using the training sets y_pred=clf.predict(X_test)
clf.fit(X_train,y_train)

y_pred=clf.predict(X_test)
```

```python
#Import scikit-learn metrics module for accuracy calculation
from sklearn import metrics
# Model Accuracy, how often is the classifier correct?
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```
